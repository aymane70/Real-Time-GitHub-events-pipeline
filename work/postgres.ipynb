```python
#[postgres_config]
host = "postgres"
port = 5432
username = "reddit_user"
password = "reddit_pass"
database = "reddit_db"
```

```python
import psycopg2 

conn = psycopg2.connect(
    "dbname=" + database
    + " user=" + username
    + " password=" + password
    + " host=" + host,
    port = port)
```

```python
# Roll back the failed transaction first
conn.rollback()

# Create a new cursor and execute the table creation
m_cursor = conn.cursor()
m_query = """
CREATE TABLE IF NOT EXISTS github_events (
    id TEXT PRIMARY KEY,
    type TEXT,
    repo_name TEXT,
    actor_login TEXT,
    created_at TIMESTAMP
)
"""
m_cursor.execute(m_query)
conn.commit()
m_cursor.close()
```
